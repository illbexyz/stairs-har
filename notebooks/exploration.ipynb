```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("../data/uci_train_6.csv")
zzz = df[df.id == 0]
```

```python
def plot_activity(label, df):
    data = df[df['label'] == label][['x', 'y', 'z']][:200]
    axis = data.plot(subplots=True, figsize=(12, 4), 
                     title=label)
    for ax in axis:
        ax.legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
```

```python
plot_activity("walking", df)
```

```python
plot_activity("stairs_down", df)
```

```python
plot_activity("stairs_up", df)
```
